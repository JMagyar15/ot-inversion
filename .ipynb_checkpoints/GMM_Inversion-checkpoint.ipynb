# Gaussian Mixture Model Inversion
## 1D 1-Component Case

We will begin with the most simple of cases; the inversion of 1-dimensional 1-component mixtures. This in itself is not a useful application but provides a way of testing some of the simple computations in a controlled environment.

```python
import numpy as np
import matplotlib.pyplot as plt
from Core_Functions import gmm_base as gb
from Core_Functions import gmm_plot as gp
from Core_Functions import optimisation as op
```

```python
#firstly, need to define a source and target distribution

source = gb.Gaussian_Mixture(1,1)
target = gb.Gaussian_Mixture(1,1)

sw = np.array([1])
sm = np.zeros([1,1])
sc = np.zeros([1,1,1])

tw = np.array([1])
tm = np.zeros([1,1])
tc = np.zeros([1,1,1])

sm[0,0] = 2
sc[0,0,0] = 1
tm[0,0] = 4
tc[0,0,0] = 5

source.assign_w(sw)
source.assign_m(sm)
source.assign_cov(sc)

target.assign_w(tw)
target.assign_m(tm)
target.assign_cov(tc)
```

```python
gp.Source_Target_1D(source,target,[-6,10])
```

```python
#now consider moving the mean of the source across the target
n = 50
mu = np.linspace(-2,10,n)
GW = np.zeros(n)
dGW_dmu = np.zeros(n)

for i in range(n):
    sm[0,0] = mu[i]
    source.assign_m(sm)
    GW[i], P, alpha = gb.GMM_Transport(source,target,0.01)
    grad = gb.Wasserstein_Gradients(source,target,P,alpha)
    dGW_dmu[i] = grad.m[0,0]

fig, ax = plt.subplots(ncols=2, figsize=[8,4])
plt.subplots_adjust(wspace=0.4)
ax[0].plot(mu,GW,'-k',linewidth=3)
ax[1].plot(mu,dGW_dmu,'-k',linewidth=3)
ax[0].set_xlabel(r'$\mu$',fontsize=14)
ax[1].set_xlabel(r'$\mu$',fontsize=14)
ax[0].set_ylabel(r'$\mathcal{GW}_2^2(f,g)$',fontsize=14)
ax[1].set_ylabel(r'$\partial\mathcal{GW}_2^2/\partial \mu$',fontsize=14)
ax[0].tick_params(labelsize = 12)
ax[1].tick_params(labelsize = 12)
```

```python
sm[0,0] = 2

source.assign_m(sm)

n = 50
cov = np.linspace(0.1,10,n)
GW = np.zeros(n)
dGW_dsigma = np.zeros(n)

for i in range(n):
    sc[0,0,0] = cov[i]
    source.assign_cov(sc)
    GW[i], P, alpha = gb.GMM_Transport(source,target,0.01)
    grad = gb.Wasserstein_Gradients(source,target,P,alpha)
    dGW_dsigma[i] = grad.cov[0,0,0]

fig, ax = plt.subplots(ncols=2, figsize=[8,4])
plt.subplots_adjust(wspace=0.4)
ax[0].plot(cov,GW,'-k',linewidth=3)
ax[1].plot(cov,dGW_dsigma,'-k',linewidth=3)
ax[0].set_xlabel(r'$\sigma^2$',fontsize=14)
ax[1].set_xlabel(r'$\sigma^2$',fontsize=14)
ax[0].set_ylabel(r'$\mathcal{GW}_2^2(f,g)$',fontsize=14)
ax[1].set_ylabel(r'$\partial\mathcal{GW}_2^2/\partial \sigma^2$',fontsize=14)
ax[0].tick_params(labelsize = 12)
ax[1].tick_params(labelsize = 12)
```

### Optimising with BFGS

```python
def Update_H(H,s,y):
    n = np.size(y)
    I = np.eye(n)
    A = I - (np.outer(s,y) / np.sum(y * s))
    B = (np.outer(s,s)/ np.sum(y * s))
    new_H = (A @ H @ A) + B
    return new_H

def BFGS(f,g,reg,num_iter=20):
    #we firstly need to initialise a few things for the first iteration
    n = f.n
    d = n * (1 + f.d + f.d**2)
    H = np.eye(d)
    
    GW = np.zeros(num_iter + 1)
    GW[0], P, alpha = gb.GMM_Transport(f,g,reg)
    grad = gb.Wasserstein_Gradients(f,g,P,alpha)
    
    for k in range(num_iter):
        # first thing to do each time: get the new gradient
        
        f_flat, d = op.Flattener(f)
        grad_flat, d = op.Flattener(grad)
        
        # this then needs to be used with the current hessian approx to get direction
        p = H @ grad_flat
        
        p_mix = op.Unflattener(p,d) #get into mixture form
        
        #now time for a line search: update the source
        a_k = op.Linear_Steps(f, p_mix, g, reg,num_steps=20)
        f_new = op.Search_Point(f,p_mix,a_k)
        
        # we now need the finite differences to update H
        f_flat_new, d = op.Flattener(f_new)
        print(f_flat_new)
        s = f_flat_new - f_flat

        
        #to get the derivates, need to do OT
        GW[k+1], P, alpha = gb.GMM_Transport(f_new,g,reg)
        print(GW[k+1])
        grad_new = gb.Wasserstein_Gradients(f_new,g,P,alpha)
        grad_flat_new, d = op.Flattener(grad_new)
        
        #now the finite difference
        y = grad_flat_new - grad_flat
        H = Update_H(H,s,y)
        
        #replace the source and gradient with the new updates
        f = f_new
        grad = grad_new
    return f, GW
```

```python
f_test, GW = BFGS(source,target,0.01,num_iter=15)
```

```python
f = gb.Gaussian_Mixture(2,2)
g = gb.Gaussian_Mixture(2,2)

f.assign_w(np.array([0.6,0.4]))
g.assign_w(np.array([0.5,0.5]))

fm = np.zeros([2,2])
gm = np.zeros([2,2])

fcov = np.zeros([2,2,2])
gcov = np.zeros([2,2,2])

fm[0,:] = np.array([-1,0])
fm[1,:] = np.array([1,1])

gm[0,:] = np.array([0.5,0.5])
gm[1,:] = np.array([1,-0.5])

fcov[0,:,:] = np.array([[1,0],[0,0.5]])
fcov[1,:,:] = np.array([[0.3,0.1],[0.1,0.5]])

gcov[0,:,:] = np.array([[0.5,0],[0,1]])
gcov[1,:,:] = np.array([[1,0],[0,0.2]])

f.assign_m(fm)
f.assign_cov(fcov)

g.assign_m(gm)
g.assign_cov(gcov)


gp.Source_Target_2D(f,g,[-2,2],[-2,2])
```

```python
# firstly, each of the step sizes that we will try
steps = np.linspace(0,1,20)
grad = gb.Wasserstein_Gradients(f,g,P,alpha)
GW = np.zeros(20)
i = 0
for a in steps:
    f_a = op.Search_Point(f,grad,a)
    GW[i] = gb.GMM_Transport(f_a,g,0.01)[0]
    i += 1
print(GW)
best = np.nanargmin(GW)
f_new = op.Search_Point(f,grad,best)
plt.plot(GW)
```
