```python
import numpy as np
from scipy.stats import multivariate_normal
from scipy.stats import multinomial
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

from Core_Functions import gmm_base as gb
from Core_Functions import gmm_plot as gp
from Core_Functions import optimisation as op
```

```python
def setup(ax):
    ax.spines['right'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.yaxis.set_major_locator(ticker.NullLocator())
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.tick_params(which='major', width=1.00)
    ax.tick_params(which='major', length=5)
    ax.tick_params(which='minor', width=0.75)
    ax.tick_params(which='minor', length=2.5)
    ax.set_ylim(0, 1)
    ax.patch.set_alpha(0.0)
```

So we can now generate some samples from an underlying standard normal distribution. These are plotted on the line below.

```python
samples = multivariate_normal(mean=0,cov=1).rvs(30)

fig, ax = plt.subplots(figsize=[8,1])
setup(ax)
ax.xaxis.set_major_locator(ticker.NullLocator())
ax.xaxis.set_minor_locator(ticker.NullLocator())
ax.scatter(samples,np.zeros(30))
```

```python
kernel_means = samples.reshape([30,1])
kernel_weights = np.full([30],1/30)
kernel_var = np.full([30,1,1],0.1)

kernel_mix = gb.Gaussian_Mixture(30,1)
kernel_mix.assign_m(kernel_means)
kernel_mix.assign_w(kernel_weights)
kernel_mix.assign_cov(kernel_var)
```

To estimate the continuous distribution, a kernel density estimation approach will be taken. Using normal distributions centred at each point with equal weights and a covariance of $0.1$. The result is displayed in the figure below. This now defines a mixture model that the devloped BFGS method can be used to fit with a 1 component mixture.

```python
gp.Source_Target_1D(kernel_mix,kernel_mix,[-5,5])
```

```python
gaus_mix = gb.Gaussian_Mixture(1,1)
gaus_mix.assign_w(np.array([1]))
gaus_mix.assign_m(np.array([[1]]))
gaus_mix.assign_cov(np.array([[[1]]]))
```

```python
fit = op.BFGS_Optimisation(gaus_mix,kernel_mix,num_iter=20)
```

```python
fit_mix = gb.Gaussian_Mixture(1,1)
fit_mix.assign_w(np.array([1]))
fit_mix.assign_m(np.array([[fit[1]]]))
fit_mix.assign_cov(np.array([[[fit[2]]]]))
```

```python
gp.Source_Target_1D(fit_mix,kernel_mix,[-5,5])
```

We can see that the fitting has aligned the distributions but has not captured the variance. This is an underlying flaw in this method of OT. As transport is acheived by shifting and reshaping the components individually, the work for each elementary movement will be minimised if the source and target components have the same variance. In the above case, the source was adjusted to have a variance near $0.1$ so that only a shifting of mean was required to transport to each kernel density estimator and thus the cost minimised.

```python
#let's try with 3 components - we expect the issue to be the same here with var=1 for all but will be
#less obvious here as the weights will decrease. Fit becomes better as components increase until exact when n=m but
#in no cases does it do a good job unless pre-defined variances are used in target that are not all equal.

gaus_mix = gb.Gaussian_Mixture(3,1)
gaus_mix.assign_w(np.array([1/3,1/3,1/3]))

mean = np.zeros([3,1])
var = np.zeros([3,1,1])

mean[:,0] = np.array([1,0.5,0])
var[:,0,0] = np.array([1,1,1])

gaus_mix.assign_m(mean)
gaus_mix.assign_cov(var)
```

```python
fit = op.BFGS_Optimisation(gaus_mix,kernel_mix,num_iter=20)
```

```python
fit_mix = op.Unflattener(fit,1)
```

```python
gp.Source_Target_1D(fit_mix,kernel_mix,[-5,5])
```
