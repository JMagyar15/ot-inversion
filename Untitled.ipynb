```python
#so there will be some weighted posterior distribution. Usually we would evaluate this by multiplying the prior and the
#likelihood but here we will just have a weighted dist.

m = np.linspace(-10,10,100)

def Evaluate_Posterior(m):
    post = (1.4 * multivariate_normal.pdf(m,mean = 0.5,cov=1.4) 
               + 0.6 * multivariate_normal.pdf(m,mean=4.4,cov=0.3))
    return post

w_posterior = Evaluate_Posterior(m)
#our prior will need to be an actual probability distribution
prior = (0.7 * multivariate_normal.pdf(m,mean=-2.5,cov=2.4) 
         + 0.3 * multivariate_normal.pdf(m,mean=6,cov=0.4))


fig, ax = plt.subplots()
ax.plot(m,w_posterior,'--',color='grey')
ax.plot(m,prior,'-k')
```

```python
#lets now make the prior a mixture
f = gb.Gaussian_Mixture(2,1)
f.assign_w(np.array([0.7,0.3]))
f.assign_m(np.array([-2.5,6]).reshape([2,1]))
f.assign_cov(np.array([2.4,0.4]).reshape([2,1,1]))

#and the inverse alpha maps will be our initial guess so lets find those
iam = mp.Inv_Alpha_Maps(f)
am = mp.Map_Inversion(iam)

#lets pick some points to make the evaluations at
m_points = mp.Model_Points(np.linspace(-10,10,100).reshape([100,1]))
post_eval = Evaluate_Posterior(m_points.all)
```

```python
def Grad_Scaling(scaling,n_points):
    """
    Inputs:
        scaling: the scaling ratio at each point from the last iteration
        n_points: the backmapped points where the scaling ratio was evaluated at
    """
    return scaling * n_points.all.T

def Kernels(scaling,n_points,Beta_Maps,j,m_points):
    a = Beta_Maps.A[j,:,:]
    b = Beta_Maps.b[j,:]
    dS_dm = Grad_Scaling(scaling,n_points).T
    dm_da = (b - m_points.all) / a**2
    dm_db = - (1 / a) * np.ones(np.shape(m_points.all))
    dS_da = dS_dm * dm_da
    dS_db = dS_dm * dm_db
    return dS_da, dS_db
```
