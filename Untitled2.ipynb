# Wasserstein Posterior Fitting

Here, we want to try and fit the posterior distribution by minimising the Wasserstein distance between our candidate mixture and the posterior. We only have evaluations of the unnormalised posterior but that is fine, as OT is better suited (computationally) for discrete distributions. We therefore want discrete approximations of the mixture and posterior. This can be done with importance sampling. Note that the discrete approximation can be made due to the stability of optimal transport (see theorem 5.20 in Villani 2009). 

## Fitting of Gaussians

```python
import numpy as np
import matplotlib.pyplot as plt
import ot
from scipy.stats import multivariate_normal
from Core_Functions import simple as sp
from Core_Functions import gmm_base as gb
```

```python
mix = gb.Gaussian_Mixture(np.array([1]),np.array([0]).reshape([1,1]),np.array([1.5]).reshape([1,1,1]))
post = gb.Gaussian_Mixture(np.array([1]),np.array([4]).reshape([1,1]),np.array([0.5]).reshape([1,1,1]))


plot_dom = sp.Model_Points(np.linspace(-5,10,100).reshape([100,1]))
fig, ax = plt.subplots()
ax.plot(plot_dom.all,sp.Evaluate_Mixture(mix,plot_dom),'-k',label='Posterior Estimate')
ax.plot(plot_dom.all,sp.Evaluate_Mixture(post,plot_dom),'--k',label='Exact Posterior')
ax.legend(fontsize=14)
```

```python
#we can now generate some samples from each at points defined 
#by the current estimate but mass distributed according to correct distribution

m_points = sp.Sample_Mixture(mix,500)

mix_eval = sp.Evaluate_Mixture(mix,m_points)
post_eval = sp.Evaluate_Mixture(post,m_points)

fig, ax = plt.subplots()
ax.plot(m_points.all,mix_eval,'or',label='Posterior Estimate')
ax.plot(m_points.all,post_eval,'ob',label='Exact Posterior')
ax.legend(fontsize=14)
```

```python
p = mix_eval / mix_eval
p /= np.sum(p)

q = post_eval / mix_eval
q /= np.sum(q)

fig, ax = plt.subplots()
ax.plot(m_points.all,p,'or',label='Posterior Estimate')
ax.plot(m_points.all,q,'ob',label='Exact Posterior')
ax.legend(fontsize=14)
```

```python
print('Computing cost matrix...')
C = ot.dist(m_points.all,m_points.all)
print('Done!')
print('Sinkhorn iterations...')
P, log = ot.bregman.sinkhorn_epsilon_scaling(p,q,C,0.01,log=True)
print('Done!')
```

```python
#so now we have the important pieces from the OT: the scaling vectors and therefore the Wasserstein approximation

W2 = np.inner(log['alpha'],p) + np.inner(log['beta'],q)
print('W2 = ',W2)
```

```python
#importantly, we also have the gradient with respect to the value at each point of the source
plt.plot(m_points.all,log['alpha'] - np.mean(log['alpha']),'ok')
```
